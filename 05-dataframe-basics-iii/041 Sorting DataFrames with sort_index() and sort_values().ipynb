```python
import pandas as pd
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic
```

## sort_values()

```python
titanic.age.sort_values()
```

```python
titanic.sort_values(by = "age")
```

```python
titanic.sort_values(by = "age", ascending = True)
```

```python
titanic.sort_values(by = "age", ascending = False)
```

```python
titanic
```

```python
titanic.sort_values(by = "age", ascending = False, inplace = True)
```

```python
titanic
```

```python
titanic.sort_values(by = ["pclass", "sex", "age"], ascending = True, inplace = True) 
```

elements in by = [...] take priority left to right, with left given first priority

```python
titanic
```

```python
titanic.sort_values(by = ["pclass", "sex", "age"], ascending = [True, True, False], inplace = True) 
```

```python
titanic
```

## sort_index()

```python
titanic.sort_index(ascending = True, inplace = True)
```

```python
titanic
```

```python
titanic.sort_values(by = "age").reset_index(drop = True)
```

```python
titanic.sort_values(by = "age", ignore_index = True)
```
