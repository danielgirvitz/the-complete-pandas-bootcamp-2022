```python
import pandas as pd
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.head(10)
```

Hypothesis: Adult men disproportionally died.

```python
mask1 = titanic.sex == "male"
mask1.head()
```

```python
mask2 = titanic.age > 14
mask2.head()
```

```python
(mask1 & mask2).head()
```

```python
male_surv = titanic.loc[mask1 & mask2, ["survived", "pclass", "sex", "age"]]
male_surv.head()
```

```python
male_surv.info()
```

```python
male_surv.describe()
```

~17% of all male passengers older than 14 yrs old survived.
