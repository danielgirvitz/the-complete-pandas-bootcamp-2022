# Selecting Rows with iloc (position-based indexing)

```python
import pandas as pd
summer = pd.read_csv("summer.csv")
summer
```

```python
summer.info()
```

#### Change index of rows

```python
summer = pd.read_csv("summer.csv", index_col = "Athlete")
summer
```

#### Selecting Rows with iloc[]

```python
type(summer.iloc[0]) 
```

```python
summer.iloc[0] # 0 is the first row
```

```python
summer.iloc[-1] # 1 is the last row
```

```python
summer.iloc[[1, 2, 3]]
```

```python
summer.iloc[1:3] # inclusive:exclusive
```

```python
summer.iloc[:5] # First 5 rows
```

```python
summer.iloc[-5:] # Last 5 rows
```

```python
summer.iloc[[2, 45, 5467]]
```

#### Indexing/Slicing Rows and Columns with iloc[] 

```python
summer.head()
```

```python
# First 10 rows
summer.head(10)
```

```python
# retrieve row#0 col#0
summer.iloc[0,0]
```

```python
# retrieve row#0 col#4
summer.iloc[0,4]
```

```python
# retrieve row#0 and first 3 cols
summer.iloc[0,:3]
```

```python
summer.iloc[0,[0,2,4,6]]
```

```python
summer.iloc[34:39,[0,2,4,6]] # inclusive:exclusive
```

#### Selecting Columns with iloc[]

```python
summer.iloc[:,4] # all rows, country col
```
