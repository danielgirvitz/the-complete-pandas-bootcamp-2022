## If you want to work with and manipulate the whole DataFrame...

## ... avoid chain indexing!!!

```python
import pandas as pd
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.head()
```

```python
titanic.iloc[1,3]=40
```

```python
titanic.head()
```

```python
index_babies = titanic[titanic.age < 1].index
```

```python
index_babies
```

```python
titanic.loc[titanic.age < 1, "age"] = 1
```

```python
titanic.loc[index_babies]
```

## If you want to work with and manipulate a Slice of a DataFrame...

## ...and makr a copy with a .copy()

#### example 01

```python
import pandas as pd
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.head()
```

```python
age = titanic.age.copy()
```

```python
age[1] = 40
```

```python
age.head()
```

```python
titanic.head()
```

#### example 02

```python
baby_ages = titanic.loc[titanic.age < 1, ["age", "sex"]].copy()
```

```python
baby_ages
```

```python
baby_ages["age"] = 1
```

```python
baby_ages
```

```python
titanic.loc[index_babies]
```
