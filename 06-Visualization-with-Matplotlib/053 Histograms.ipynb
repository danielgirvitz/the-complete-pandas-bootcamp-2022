```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8')
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.head()
```

```python
titanic.age.value_counts()
```

```python
titanic.age.value_counts(bins = 10)
```

```python
titanic.age.plot(kind = "hist", figsize = (8,6), fontsize = 15)
plt.show()
```

```python
titanic.age.plot(kind = "hist", figsize = (8,6), fontsize = 15, bins = 30)
plt.show()
```

```python
titanic.age.hist(figsize = (8,6), bins = 80, xlabelsize = 15, ylabelsize = 15)
plt.show()
```

```python
plt.figure(figsize = (8,6))
plt.hist(titanic.age, bins = 80, density = False, cumulative = False)
plt.show()
```

```python
plt.figure(figsize = (8,6))
plt.hist(titanic.age, bins = 80, density = False, cumulative = True)
plt.show()
```

```python
plt.figure(figsize = (8,6))
plt.hist(titanic.age, bins = 80, density = True, cumulative = True)
plt.show()
```
