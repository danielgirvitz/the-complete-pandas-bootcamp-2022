```python
import pandas as pd
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
import matplotlib.pyplot as plt
```

```python
titanic.age.plot(figsize = (8,6))
plt.show()
```

```python
titanic.age.plot(figsize = (8,6),
                 fontsize = 13,
                 c = "r",
                 linestyle = "--")
plt.show()
```

```python
titanic.age.plot(figsize = (8,6),
                 fontsize = 13,
                 c = "r",
                 linestyle = "-")
plt.title("Titanic - Ages", fontsize = 15)
plt.legend(loc = "best", fontsize = 15)
plt.xlabel("Passenger No", fontsize = 13)
plt.ylabel("Age", fontsize = 13)
plt.grid()
plt.show()
```

```python
titanic.age.plot(figsize = (8,6),
                 fontsize = 13,
                 c = "r",
                 linestyle = "-",
                 xlim = (200,400))
plt.title("Titanic - Ages", fontsize = 15)
plt.legend(loc = "best", fontsize = 15)
plt.xlabel("Passenger No", fontsize = 13)
plt.ylabel("Age", fontsize = 13)
plt.grid()
plt.show()
```

```python
xticks = [x for x in range(0,901,50)]
xticks
```

```python
yticks = [y for y in range(0,81,5)]
yticks
```

```python
titanic.age.plot(figsize = (8,6),
                 fontsize = 13,
                 c = "r",
                 linestyle = "-",
                 xlim = (200,400),
                 xticks = xticks,
                 yticks = yticks)
plt.title("Titanic - Ages", fontsize = 15)
plt.legend(loc = "best", fontsize = 15)
plt.xlabel("Passenger No", fontsize = 13)
plt.ylabel("Age", fontsize = 13)
plt.grid()
plt.show()
```

```python
titanic.age.plot(figsize = (8,6),
                 fontsize = 13,
                 c = "r",
                 linestyle = "-",
                 xlim = (200,400),
                 xticks = xticks,
                 yticks = yticks,
                 rot = 45)
plt.title("Titanic - Ages", fontsize = 15)
plt.legend(loc = "best", fontsize = 15)
plt.xlabel("Passenger No", fontsize = 13)
plt.ylabel("Age", fontsize = 13)
plt.grid()
plt.show()
```
