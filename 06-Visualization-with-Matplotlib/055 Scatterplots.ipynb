```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8')
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.head()
```

```python
titanic.plot(kind = "line", figsize = (8,6))
plt.show()
```

```python
titanic.plot(kind = "scatter", figsize = (8,6), x = "age", y = "fare")
plt.show()
```

```python
titanic.plot(kind = "scatter", figsize = (8,6), x = "age", y = "fare", c = "red", marker = "o")
plt.show()
```

```python
titanic.plot(kind = "scatter", figsize = (8,6), x = "age", y = "fare", c = "red", marker = "o", s = 10)
plt.show()
```

```python
titanic.plot(kind = "scatter", figsize = (8,6), x = "age", y = "fare", c = "survived", marker = "o")
plt.show()
```

```python
titanic.plot(kind = "scatter", figsize = (8,6), x = "age", y = "fare", c = "survived", marker = "o", colormap = "viridis")
plt.show()
```

```python
titanic.plot(kind = "scatter", figsize = (8,6), x = "age", y = "fare", c = "pclass", marker = "o", colormap = "viridis")
plt.show()
```
